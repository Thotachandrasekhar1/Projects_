```python
Rank = []
ChannelName = []
Subscribers = []
Country = []
Views = []
Likes = []
Comments = []

for i in range(1,11):
    urls = "https://hypeauditor.com/top-youtube/?p="+str(i)
    print(urls)
    page = requests.get(urls)
    soup = BeautifulSoup(page.text)
 
    # rank
    a = soup.find_all("div",class_ = "row-cell rank")
    for i in a:
    
        Rank.append(i.text)

    # Channelnames
    b = soup.find_all("div",class_ = "contributor__name-content")

    for i in b:
        a = i.text
        if a == '':
            ChannelName.append(np.nan)
        else:
            ChannelName.append(i.text)
    
     # Catogery
        c = soup.find_all("div",class_ = "contributor__title")
    for i in c:
        Category.append(i.text)
        # subscribers
    d = soup.find_all("div",class_ = "row-cell subscribers")
    for i in d:
    
        Subscribers.append(i.text)
        
    # country    
    e = soup.find_all("div",class_ = "row-cell audience")
    for i in e:
        a = i.text
        
        if a == '':
            Country.append(np.nan)
        else:
            Country.append(a)
            
    # views        
    f = soup.find_all("div",class_ = "row-cell avg-views")
    for i in f:
       
         Views.append(i.text)
        
    # likes    
    g = soup.find_all("div",class_ = "row-cell avg-likes")
    for i in g:
        
        Likes.append(i.text)
        
    # comments    
    h = soup.find_all("div",class_ = "row-cell avg-comments")
    for i in h:
        
        Comments.append(i.text)
```

```python
Category = Category[0:500]
```

```python
len(ChannelName),len(Likes),len(Comments),len(Subscribers),len(Country),len(Rank),len(Views),len(Category)
```

```python
df={"Rank":Rank,
   "Channel":ChannelName,
   "Category":Category,
    "Country":Country,
   "Likes":Likes,
   "Comments":Comments,
   "Views":Views,
   "Subscribers":Subscribers}
```

```python
# Exclude rows with missing values in 'Country' or 'Subscribers' columns
df_filtered = df.dropna(subset=['Country', 'Subscribers'])

# Convert the 'Subscribers' column to string
df_filtered['Subscribers'] = df_filtered['Subscribers'].astype(str)

# Remove 'M' from 'Subscribers' column and convert to float
df_filtered['Subscribers'] = df_filtered['Subscribers'].str.replace('M', '').astype(float)

# Group by 'Country' and sum the 'Subscribers' column
country_subscribers = df_filtered.groupby('Country')['Subscribers'].sum().reset_index()

# Sort the DataFrame by 'Subscribers' column in descending order
country_subscribers_sorted = country_subscribers.sort_values('Subscribers', ascending=False).head(10)

# Create the bar chart using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Subscribers', y='Country', data=country_subscribers_sorted)

# Set the plot title and axis labels
plt.title('Top 10 Countries by Subscriber Count')
plt.xlabel('Subscriber Count (in millions)')
plt.ylabel('Country')

# Display the plot
plt.show()
```

* Indian YouTube channels have significantly higher subscribers, views, likes, and comments than non-Indian channels.
